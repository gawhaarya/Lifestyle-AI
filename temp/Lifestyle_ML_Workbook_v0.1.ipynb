```python
from datetime import date
current_year = date.today().year
data['chronological_age'] = current_year - data['BirthYear']
```

```python
data['biological_age'] = data.iloc[:,1:12].sum(axis=1)
```

```python
data['temp'] = data['biological_age'] - data['chronological_age']
```

```python
classes = []
for i in data['temp']:
    if(i==0):
        classes.append('Equals')
    elif (i<0):
        classes.append('Outstanding')
    elif (i>0 and i<=5):
        classes.append('At low risk')
    elif(i>5 and i<20):
        classes.append('At risk')
    else:
        classes.append('At major risk')
print(len(classes))
#print(classes)
data['LifeStyle_Score']=classes
```
