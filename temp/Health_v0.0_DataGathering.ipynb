```python
#classes = ['Outstanding','Equals','At risk','At major risk']
#negative,equal,positive,lagre positive
```

```python
dataset['Temp']=dataset['Biological_Age']-dataset['Chronological_Age']
```

```python
classes = []

for i in dataset['Temp']:
    if(i==0):
        classes.append('Equals')
    elif (i<0):
        classes.append('Outstanding')
    elif (i>0 and i<=5):
        classes.append('At low risk')
    elif(i>5 and i<20):
        classes.append('At risk')
    else:
        classes.append('At major risk')
print(len(classes))
print(classes)
```

```python
dataset['LifeStyle_Score']=classes
```
